# 데이터선택(조건)
조건에 해당하는 데이터 선택해서 가져오기

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
#키가 185이상인지 확인하고싶다
df['키']>=185
```

```python
#키 185이상인 항목만 다 들고오기.  2번,3번이 빠지게된다
filt=(df['키']>185)  #조건설정
df[filt]
```

```python
df[-filt]                #역필터
```

```python
df[df['키']>185]  # =df[filt]
```

```python
#키가 185이상인 학생들의 수학점수가 궁금하다
df.loc[df['키']>185,'수학']
```

```python
df.loc[df['키']>185, ['이름','수학','과학']]
```

## 다양한 조건


### 그리고&

```python
df.loc[(df['키']>=185)&(df['학교']=='북산고')]     #키가 185이상,북산고 인 학생
```

### 또는 |      (shift백스페이스와 엔터사이의 원화)

```python
#키가 너무 작거나 너무 큰 친구만 가져와보자.
df.loc[(df['키']<=170)|(df['키']>=200)]
```

### str함수

```python
filt=df['이름'].str.startswith('송') #송으로 시작하는 이름을 가진 데이터정보
df[filt]
```

```python
filt=df['이름'].str.contains('태') #태가 이름에 들어가 있는 사람
df[filt]
```

```python
df[-filt]
```

```python
langs=['Python','Java']
filt=df['SW특기'].isin(langs)
df[filt]
#허나 이렇게 하면 PYTHON대문자를 가진 황태산은 안나온다.
```

```python
langs=['python','java']
filt=df['SW특기'].str.lower().isin(langs)  #lower함수추가로 대소문자동일취급
df[filt]
```

```python
#java라는 글자가 들어가는 SW특기 가진 사람 다 뽑기
filt=df['SW특기'].str.contains('Java',na=False)  #na=False는 왜 필요한가? 비어있는 칸은 NA칸인데 이것이 있으면contains함수가 filt가 오류가난다. na는 미표기라고 알려주는 행위이다.
df[filt]
```
