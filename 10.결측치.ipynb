## 데이터 제외하기 dropna

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.dropna(inplace=True) #na가 포함된 데이터 아예 삭제
df
```

axis:  index or colums      #지워지는데이터가 index(row)기준인지 colums기준인지 선택하기  
how: any or all             #NA가 하나라도 있으면 지우기 or 다 na면 지우기


```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
df.dropna(axis='index',how='any',inplace=True)
df
```

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
df.dropna(axis='columns')    #how는 any가 default값이다.
```

```python
import numpy as np   #numpy는 범위외. 일단 무시
df['학교']=np.nan     #학교 데이터 전체를 NaN으로 채움
df
```

```python
df.dropna(axis='columns',how='all',inplace=True)
df
#columns데이터 전체가 na일 경우만 데이터삭제
```
